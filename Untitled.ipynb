```python
from LAC import LAC
import pandas as pd
import click
import genanki
from chinese_english_lookup import Dictionary
```

```python
lac = LAC(mode='seg')
text = u"俄羅斯瓦格納集團首腦今天宣布同意撤軍，以避免「流血事件」，並緩和對總統普亭構成嚴重挑戰的局勢；普亭則對居中斡旋撤軍的白俄總統表達感謝之意。路透社報導，瓦格納集團首腦、曾是普亭忠實盟友的普里格津（Yevgeny Prigozhin）表示，截至撤軍之前，瓦格納戰士距離首都僅有200公里。他的部隊已佔領俄國南方的羅斯托夫市（Rostov），原本正準備進軍莫斯科。普里格津在他發布的語音訊息中說道，「他們想要解散瓦格納軍事連。我們在6月23日發起了『正義遊行』，24小時之內，我們就到達了距離莫斯科200公里以內的地方。截至此時此刻，我們的戰士沒有流一滴血。」「現在，可能流血的時刻已經到來。因為了解到…某一方可能流下俄羅斯人的鮮血，我們正在調轉縱隊，按計畫返回營地。」白俄羅斯總統魯卡申柯（Alexander Lukashenko）的辦公室表示，瓦格納集團停止在俄羅斯進一步行動的決定，是由盧卡申科居中斡旋，以換取莫斯科對瓦格納戰士的安全保證"
articleName = "Wagner and Putin"
seg_result = lac.run(text)
print(seg_result)
```

```python
nonRepeat = []
badChar = ['，', ', ', '。', ' ','「','」','「','『','、','（',')','…','，']
for word in badChar:
    if seg_result.count(word) > 0:
        seg_result.remove(word)
    
for word in seg_result:
    if nonRepeat.count(word) < 1:
        nonRepeat.append(word)
        
print(nonRepeat)
```

```python
d = Dictionary()

keepQstn = []
keepTrans = []
keepPinYin = []
noDefFound = []
for word in nonRepeat:
    word_entry = d.lookup(word)
    if word_entry is None:
        noDefFound.append(word)
        print('No Definition found for' + word)
        nonRepeat.remove(word)
        
for word in nonRepeat:
    word_entry = d.lookup(word)
    if click.confirm(f"{word_entry} ||| Include this entry?", default=True):
        keepQstn.append(word)
        keepTrans.append(word_entry.definition_entries[0].definitions[0])
        keepPinYin.append(word_entry.definition_entries[0].pinyin)
        
```

```python
print(keepQstn)
print(keepTrans)
print(keepPinYin)
```

```python
i = 0
my_deck = genanki.Deck(
2059400110,
articleName)

while i < len(keepQstn):    
    my_model = genanki.Model(
        1607392319,
        'Simple Model',
        fields=[
            {'name': 'Question'},
            {'name': 'Answer'},
        ],
        templates=[
            {
                'name': i,
                'qfmt': '{{Question}}',
                'afmt': '{{FrontSide}}<hr id="answer">{{Answer}}',
            },
        ])

    my_note = genanki.Note(
        model=my_model,
        fields=[f"{keepQstn[i]}", f"{keepPinYin[i]} <hr> {keepTrans[i]}"]
        )
    my_deck.add_note(my_note)
    print("Created note for " + keepQstn[i] + "  -  " + keepTrans[i])
    i = i + 1

print(f"{len(keepQstn)} new cards generated!")
    
```
